```python
import cv2
import numpy as np
from PIL import Image
from keras import models
import os
import tensorflow as tf
```

```python
checkpoint_dir = os.path.dirname(current_path)
```

```python
model = models.load_model('CNN_Model.h5')
```

```python
video = cv2.VideoCapture(0)
while True:
        temp = []
        _, frame = video.read()

        #Convert the captured frame into RGB
        im = Image.fromarray(frame, 'RGB')
        im = im.resize((90,90))
        img_array = np.array(im)
        img_array = img_array.reshape(1,90,90,3)
        
        
        prediction = model.predict(img_array)
        if(np.argmax(prediction) == 0 and prediction[0][np.argmax(prediction)]>0.5):
            print(prediction)
            print("warning_children")
        elif(np.argmax(prediction) == 1  and prediction[0][np.argmax(prediction)]>0.5):
            print(prediction)
            print("Turn_left")
        elif(np.argmax(prediction) == 2 and prediction[0][np.argmax(prediction)]>0.5):
            print(prediction)
            print("Turn_right")
        else:
            print('Nothing')

        cv2.imshow("Prediction", frame)
        key=cv2.waitKey(0)
        if key == ord('q'):
                break
video.release()
cv2.destroyAllWindows()
```
